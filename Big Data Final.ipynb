### Средствами pandas соедините данные по оценкам с данными по пользователям (u.user)

### Построить модель, которая предсказывает оценку для фильма и пользователя 
### (воспользоваться RandomForestRegressor, тк мы решаем задачу регрессии)

```python
user_ratings['is male'] = user_ratings['gender'] == 'M'
```

```python
user_ratings.head()
```

```python
user_ratings_f = user_ratings[['user id','age','is male','occupation','item id','rating']]
user_ratings_f.head()
```

```python
user_ratings_f = pd.get_dummies(user_ratings_f, ['occupation'])
```

```python
user_ratings_f.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

from sklearn.ensemble import RandomForestRegressor
```

```python
user_ratings_f.columns
```

```python
X, y = user_ratings_f[['user id', 'age', 'is male', 'item id','occupation_administrator', 
                       'occupation_artist', 'occupation_doctor','occupation_educator', 
                       'occupation_engineer', 'occupation_entertainment', 'occupation_executive',
                       'occupation_healthcare', 'occupation_homemaker', 'occupation_lawyer',
                       'occupation_librarian', 'occupation_marketing', 'occupation_none',
                       'occupation_other', 'occupation_programmer', 'occupation_retired',
                       'occupation_salesman', 'occupation_scientist', 'occupation_student',
                       'occupation_technician', 'occupation_writer']], user_ratings_f['rating']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
X_train = pd.DataFrame(X_train, columns=X.columns)
X_test = pd.DataFrame(X_test, columns=X.columns)
```

```python
X_train.head()
```

```python
regressor = RandomForestRegressor(n_estimators=100, random_state=0)
regressor.fit(X_train, y_train)
```

```python
regressor.predict(X_test)
```

### Оценить качество регрессора на отложенной (train_test_split) выборке

```python
from sklearn.metrics import r2_score
```

```python
r2_score (y_test, regressor.predict(X_test))
```

```python
from sklearn.metrics import mean_absolute_error
from math import sqrt

RMAE = sqrt(mean_absolute_error (y_test, regressor.predict(X_test)))
RMAE
```
